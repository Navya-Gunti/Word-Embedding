## **Using Sklearn Kmeans cluster function to cluster the created vectors**

```python
X = model1[model1.wv.vocab]
NUM_CLUSTERS=10
```

```python
import numpy as np
from sklearn import cluster
from sklearn import metrics
kmeans = cluster.KMeans(n_clusters=NUM_CLUSTERS)
kmeans.fit(X)
 
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
 
print ("Cluster id labels for inputted data")
print (labels)
```

## **Grouping the words that are in the same embedding space**

```python
for i in range(10):
  #model1.wv.most_similar(kmeans.cluster_centers_[i])
  
  print(model1.wv.most_similar([kmeans.cluster_centers_[i]]))

  #model1.most_similar([kmeans.cluster_centers_[i]]) 
  #model1.most_similar([kmeans.cluster_centers_[i]])
```

```python
#finding the centroids
print ("Centroids data")
print (centroids)
```

```python
print ("Score (Opposite of the value of X on the K-means objective which is Sum of distances of samples to their closest cluster center):")
print (kmeans.score(X))
```

```python
silhouette_score = metrics.silhouette_score(X, labels, metric='euclidean')
 
print ("Silhouette_score: ")
print (silhouette_score)
```
